```python
#Request authentication 
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
#Define tweetSearch
def tweetSearch(query, limit = 100, language = "id"):
    text = ""
    for tweet in tweepy.Cursor(api.search, q=query, lang=language).items(limit):
        text += tweet.text.lower()
    return text
```

```python
#Define wordcloud plot
def plot_cloud(wordcloud):
    #Change resolution of plotted images (20,10)= 400kb, (40,20)=1000kb  
    plt.figure(figsize=(20,10))
    plt.imshow(wordcloud, interpolation = "bilinear")
    #Show ruler on the side of the plot results
    plt.axis("off")
```
